# Local test

This notebook calls your local Azure Functions HTTP endpoint:

- **Function**: `IngestHistoricalResults`
- **URL**: `http://localhost:7071/api/IngestHistoricalKaggleResults`

- **Function**: `IngestRugby365ResultsFunction`
- **URL**: `http://localhost:7071/api/IngestRugby365ResultsFunction`
Make sure you have your Functions host running in a terminal with:

```bash
func start
```

Then run the code cell below to trigger the ingestion and inspect the response.


````python
import requests
import json

BASE_URL = "http://localhost:7071" 
endpoint = f"{BASE_URL}/api/IngestHistoricalKaggleResults"
# endpoint = "https://func-rugbybot-dev-g9f5gygqhgcaepe4.southafricanorth-01.azurewebsites.net/api/IngestHistoricalKaggleResults"


print(f"Calling {endpoint} ...")
response = requests.post(endpoint)

print("Status:", response.status_code)
try:
    data = response.json()
    print("Response JSON:", json.dumps(data, indent=2))
except ValueError:
    print("Response text:", response.text)
````
